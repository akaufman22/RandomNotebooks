```python
import re
```

```python
#text from Mike X Cohen linear algebra lecture
with open('captions_text.vtt') as f:
    text = f.read()
```

```python
print(text)
```

```python
patterns2replace = "\\n\d{1,3}\\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\\n"

regex_lines = re.compile(patterns2replace)
regex_lines.findall(text)
```

```python
newtext = regex_lines.sub('', text)
newtext = newtext.replace('\n', ' ')
print(newtext)
```

```python
with open('./newtext.txt', 'w') as f:
    f.write(newtext)
```
